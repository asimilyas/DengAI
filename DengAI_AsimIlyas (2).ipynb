```python
# Split the data into features and target label
total_cases = train_data ['total_cases']
features = train_data.drop(['total_cases', 'week_start_date'], axis = 1)
```

```python
features.head(1)
```

```python
# Split the 'features' and 'total_cases' data into training and testing sets ratio 90/10 
X_train, X_test, y_train, y_test = train_test_split(features, total_cases, test_size = 0.2851, random_state = 0)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
lr = LinearRegression()

# Train the model using the training sets and check score
lr.fit(X_train, y_train)

#Predict Output
y_pred_lr= lr.predict(X_test)
score_lr= mean_absolute_error(y_test, y_pred_lr)

print(score_lr)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Create the parameters list for Linear Regression Model

parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}


# Perform grid search cross validation
grid_obj = GridSearchCV(estimator=lr, param_grid= parameters, cv=5)

# Fit the grid search object to the training data 
grid_fit = grid_obj.fit(X_train, y_train)

# Get the estimator
best_lr = grid_fit.best_estimator_

# Make predictions using the optimized model
predictions_lr = (lr.fit(X_train, y_train)).predict(X_test)
best_predictions_lr = best_lr.predict(X_test)

print("\nOptimized Model\n------")
print("Final mean absolute error on the testing data (RF): {:.4f}".format(mean_absolute_error(y_test, best_predictions_lr)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 100)

rf.fit(X_train, y_train)

#Predict Output
y_pred_rf= rf.predict(X_test)
score_rf= mean_absolute_error(y_test, y_pred_rf)

print(score_rf)
```

```python
# Create the parameters list for RandomForest Regressor
parameters =  {
    'n_estimators': [100, 200]
}


# Perform grid search cross validation
grid_obj = GridSearchCV(estimator=rf, param_grid= parameters, cv=5)

# Fit the grid search object to the training data 
grid_fit = grid_obj.fit(X_train, y_train)

# Get the estimator
best_rf = grid_fit.best_estimator_

# Make predictions using the optimized model
predictions_rf = (rf.fit(X_train, y_train)).predict(X_test)
best_predictions_rf = best_rf.predict(X_test)

print("\nOptimized Model\n------")
print("Final mean absolute error on the testing data (RF): {:.4f}".format(mean_absolute_error(y_test, best_predictions_rf)))
```
